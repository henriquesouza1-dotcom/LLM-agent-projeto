```python
from transformers import pipeline

# Carregando um modelo pequeno da HuggingFace
llm = pipeline(
    "text-generation",
    model="google/flan-t5-base",
    max_new_tokens=200
)
```

```python
# Criando o arquivo base.txt utilizado pelo agente
with open("baseLLM.txt", "w", encoding="utf-8") as f:
    f.write("""
    Inteligência artificial é um campo da ciência da computação que busca criar máquinas capazes
    de executar tarefas que normalmente exigem inteligência humana.

    LLMs (Large Language Models) são modelos de linguagem treinados em grandes quantidades de texto.
    Eles podem responder perguntas, gerar texto, auxiliar em análises e muito mais.

    Agentes LLM combinam modelos de linguagem com ferramentas externas, como cálculos, buscas
    e ações específicas.
    """)
```

```python
# Ferramenta de cálculo
def ferramenta_calculo(expr):
    try:
        return str(eval(expr))
    except:
        return "Erro ao calcular."

# Ferramenta de busca em arquivo
def ferramenta_busca(termo, caminho="baseLLM.txt"):
    try:
        with open(caminho, "r", encoding="utf-8") as f:
            texto = f.read()

        termo_lower = termo.lower()

        if termo_lower in texto.lower():
            idx = texto.lower().find(termo_lower)
            trecho = texto[idx:idx+200]
            return f"Termo encontrado:\n...{trecho}..."
        else:
            return "Termo não encontrado no arquivo."

    except FileNotFoundError:
        return "Arquivo base.txt não encontrado."
```

```python
def agente(prompt):
    """
    Agente simples que decide automaticamente
    qual ferramenta usar (cálculo, busca ou LLM).
    """

    prompt_lower = prompt.lower()

    # Condição para cálculos
    if "calcule" in prompt_lower:
        expr = prompt_lower.replace("calcule", "").strip()
        return "Resultado: " + ferramenta_calculo(expr)

    # Condição para buscas
    if "procure por" in prompt_lower or "procure" in prompt_lower:
        termo = prompt_lower.replace("procure por", "").replace("procure", "").strip()
        return ferramenta_busca(termo)

    # Caso contrário: responder usando o LLM
    resposta = llm(prompt)[0]["generated_text"]
    return resposta
```

```python
print("Teste de cálculo:")
print(agente("Calcule 25 * 4"))
print()

print("Teste de busca:")
print(agente("Procure por inteligência artificial"))
print()

print("Teste com LLM:")
print(agente("Explique o que é um LLM Agent"))

print("Bruno Dias , Caio Limonge, Enzo Gonzalez, Éder Junior, Henrique De Oliveira, Gustavo Mori")
```
